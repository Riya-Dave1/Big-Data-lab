```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL = 'https://www.amazon.in/s?k=books&crid=1A7AO5STP7V9L&qid=1706706276&sprefix=books%2Caps%2C202&ref=sr_pg_1'
```

```python
HEADER = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3', 'Accept-Language' : 'en-US, en;q=0.5'})
```

```python
Webpage = requests.get(URL, headers=HEADER)
```

```python
Webpage
```

```python
type(Webpage.content)
```

```python
soup = BeautifulSoup(Webpage.content, 'html.parser')
```

```python
soup
```

```python
links = soup.find_all("span", class_='a-size-medium a-color-base a-text-normal')
```

```python
links
```

```python
product_names = [tag.text.strip() for tag in links]
```

```python
len(links)
```

```python
for name in product_names:
  print(name)
```

```python
len(links)
```

```python
link1 = soup.find_all("span", class_='a-icon-alt')
```

```python
link1
```

```python
link1 = soup.find_all("span", class_='a-icon-alt')
product_ratings = [tag.text.strip() for tag in link1]
for name in product_ratings:
  print(name)
```

```python
len(link1)
```

```python
link2 = soup.find_all("span", class_='a-size-base s-underline-text')
product_reviews = [tag.text.strip() for tag in link2]
for name in product_reviews:
  print(name)
```

```python
len(link2)
```

```python
link3 = soup.find_all("span", class_='a-price-whole')
product_price = [tag.text.strip() for tag in link3]
for name in product_price:
  print(name)
```

```python
len(link3)
```

```python
link4 = soup.find_all("span", class_="a-size-base a-color-secondary a-text-normal")
product_date = [tag.text.strip() for tag in link4]
for name in product_date:
  print(name)
```
