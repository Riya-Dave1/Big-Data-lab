```python
import requests

url = "https://rest-countries10.p.rapidapi.com/countries"

headers = {
	"X-RapidAPI-Key": "5e19d06647msh20daa5d4d74b830p15259fjsne48bac30167a",
	"X-RapidAPI-Host": "rest-countries10.p.rapidapi.com"
}

response = requests.get(url, headers=headers)

print(response.json())
```

```python
data = response.json()
data
```

```python
fullnames = [country.get('name', {}).get('fullname') for country in data]
# Print the extracted fullnames
for i in fullnames:
  print(i)
```

```python
shortnames = [country.get('name', {}).get('shortname') for country in data]
# Print the extracted shortnames
for i in shortnames:
  print(i)
```

```python
codes = [(country['code'].get('aplha2code'), country['code'].get('alpha3code')) for country in data]
# Print the extracted codes
for i in codes:
  print(i)
```

```python
independent_statuses = [country.get('independent') for country in data]
# Print the extracted independent statuses
for i in independent_statuses:
  print(i)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'Name': shortnames, 'Code': codes, 'Full_Name': fullnames, "Independance": independent_statuses})
df
```

```python
df.to_csv('CountryData.csv',index=False)
```
